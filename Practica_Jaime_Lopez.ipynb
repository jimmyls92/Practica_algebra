Deberas crear tu propia funci칩n para calcular el error de reconstrucci칩n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
### TODO: Funci칩n para calcular el error de reconstrucci칩n
### TU RESPUESTA ABAJO

import numpy as np
from numpy.linalg import norm

def sse_score(X, X_hat):
    """
    Funci칩n para calcular el error de reconstrucci칩n
    
    Argumentos:
        X -- Matriz Original
        X_hat -- Matriz Reconstruida
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
        sse = sse_score(X, X_hat) # -> 0.09879
    """
    # TODO
    sse = np.sum((norm(X-X_hat))**2)
    return  sse
```

```python
X = np.array([[1, 2], [3, 4]])
X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
sse = sse_score(X, X_hat) # -> 0.09879
sse
```

Una vez que ya tenemos la funci칩n `sse` hecha, podemos pasar a construir la funci칩n que ejecutar치 `SVM`.

```python
### TODO: Funci칩n para ejecutar SVM
### Tiene como entrada una matriz X
### Devuelve U, s, Vt

### Hint: S debe ser una matriz diagonal
### TU RESPUESTA ABAJO

def svm(X):
    """
    Funci칩n que ejecuta SVM y devuelve U, S, Vt
    
    Argumentos:
        X -- Matriz Original
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        U, S, Vt = svm(X)  
        
        # -> U = array([[-0.40455358, -0.9145143 ],
        #               [-0.9145143 ,  0.40455358]])
        # -> S = array([[5.4649857 , 0.        ],
        #               [0.        , 0.36596619]])   
        # -> Vt = array([[-0.57604844, -0.81741556],
        #                [ 0.81741556, -0.57604844]])          
    """
    # TODO
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    s = np.diag(s)
    
    
    return U, s, Vt
```

```python
U, s, Vt = svm(X)
print(U)
print(s)
print(Vt)
```

Como hemos visto en clase, las matrices obtenidas a partir de `SVM` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci칩n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
### TODO: Funci칩n para reconstruir la matriz original a partir de U, s, Vt
### Tiene como entrada U, s, Vt
### Devuelve X_hat
### TU RESPUESTA ABAJO


def reconstruction(U, S, Vt):
    """
    Funci칩n que reconstruye la matriz original a partir de U, s, Vt
    
    Argumentos:
        U -- Matriz de Singular Vectors
        s -- Matriz de Eigenvalues
        Vt -- Matriz de Singular Vectors
        
    Ejemplo:
        U = np.array([[-0.40455358, -0.9145143 ],
                      [-0.9145143 ,  0.40455358]])
        S = np.array([[5.4649857 , 0.        ],
                      [0.        , 0.36596619]])
        Vt = np.array([[-0.57604844, -0.81741556],
                       [ 0.81741556, -0.57604844]])
        X_hat = reconstruction(U, S, Vt)
        
        # X_hat -> array([[0.99999999, 1.99999998],
        #                 [3.00000003, 4.00000001]])
    """
    # TODO

    X_hat = U.dot(S.dot(Vt))
    
    return X_hat
```

```python
X_hat = reconstruction (U,s,Vt)
print(X_hat)
```

Calcula el error de reconstrucci칩n usando la funci칩n `sse` que has programado anteriormente.

```python
sse = sse_score(X, X_hat)
print(f"El error de reconstrucci칩n es: {sse}")
```

#### M칤nimos Cuadrados / Least Squares

Ahora, implementar치s la ecuaci칩n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
### TODO: Funci칩n para resolver la ecuaci칩n wLS
### Toma como argumentos dos matrices, una para X y otra para y
### Asumimos que las matrices tienen las dimensiones correctas

### Paso 1: Asegurate que n > d. 
### Es decir, que el n칰mero de observaciones es mayor que el n칰mero de dimensiones.
### O lo que es lo mismo, que el n칰mero de filas de cada matriz sea mayor que el n칰mero de columnas
### Si no es as칤, debes transponer las matrices

### Paso 2: Debes a침adir a la matriz X un vector columna del tama침o (n x 1)

### Paso 3: Usa la ecuaci칩n de arriba para obtener wLS

### TU RESPUESTA ABAJO


def least_squares_weights(input_x, target_y):
    """Resuelve la ecuaci칩n para wLS
    
    Argumentos:
        input_x -- Matriz con los datos de entrenamiento
        target_y -- Vector con los datos de salida
        
    Ejemplo:
        import numpy as np
        training_y = np.array([[208500, 181500, 223500, 
                                140000, 250000, 143000, 
                                307000, 200000, 129900, 
                                118000]])
        training_x = np.array([[1710, 1262, 1786, 
                                1717, 2198, 1362, 
                                1694, 2090, 1774, 
                                1077], 
                               [2003, 1976, 2001, 
                                1915, 2000, 1993, 
                                2004, 1973, 1931, 
                                1939]])
        weights = least_squares_weights(training_x, training_y)
        
        print(weights)  #--> np.array([[-2.29223802e+06],
                        #              [ 5.92536529e+01],
                        #              [ 1.20780450e+03]])
                           
        print(weights[1][0]) #--> 59.25365290008861
    
    Asumimos:
        -- target_y es un vector con el mismo n칰mero de observaciones que input_x
    """
    
    # TODO

    if np.shape(input_x)[0] < np.shape(input_x)[1]:
        input_x = input_x.transpose()
        target_y = target_y.transpose()
        ones = np.ones((np.shape(input_x)[0], 1))
        input_x = np.concatenate((ones, input_x), 1)

    else:
        print("Aqui")
        ones = np.ones(1, np.shape(input_x)[1])
        input_x = np.concatenate((ones, input_x), 0)
        
    
    w_LS = np.dot(np.dot(np.linalg.inv(np.dot(input_x.transpose(),input_x)),input_x.transpose()),target_y)
    
    return w_LS
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matem치ticos detr치s del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresi칩n.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m치ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi칩n lineal, se basa en resolver esta ecuaci칩n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

Una vez construida nuestra funci칩n para el Gradient Descent podemos usarla para encontrar los valores optimos de $w$. **Prueba a modificar el learning rate para ver la convergencia del Gradient Descent.**

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Funci칩n para convertir un conjunto a vector
    
    Argumentos:
        tokens_web1 -- Contenido tokenizado de p치gina web 1
        tokens_web_2 -- Contenido tokenizado de p치gina web 2
        set_web1 -- Conjunto de palabras de la p치gina web 1
        set_web2 -- Conjunto de palabras de la p치gina web 2
        
    Ejemplo:
        tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
        tokens_web_2 = ["hola", "me", "llamo"]
        set_web1 = {"hola", "que", "tal"}
        set_web2 = {"hola", "me", "llamo"}
        union_ls = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
        
    Requerimientos:
        Depende de la funci칩n `union()`
    """
    
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    web1_array = []
    web2_array = [] 

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return web1_array, web2_array

tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
tokens_web_2 = ["hola", "me", "llamo"]
set_web1 = {"hola", "que", "tal", "soy", "Alex"}
set_web2 = {"hola", "me", "llamo"}
web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
```
